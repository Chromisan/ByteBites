### **æ³¨æ„**

- è¯»å–åº—é“ºæ•°æ®çš„ä»£ç æ›´æ”¹ï¼Œå¯¼è‡´csvæ–‡ä»¶ä¸­å­—æ®µå‘ç”Ÿå˜åŒ–æ—¶ï¼šdef content_func ä¸­ python content_fields çš„å€¼è¦éšä¹‹æ›´æ”¹ã€‚
- æƒ³ä½œä¸ºç¡¬æ€§æŒ‡æ ‡çš„å­—æ®µå‘ç”Ÿå˜åŒ–æ—¶ï¼šè°ƒç”¨éƒ¨åˆ† metadata_fields ä¸­çš„å€¼è¦æ›´æ”¹ã€‚

### â­ æ¨èç»“æœå¡ç‰‡å±•ç¤ºä¸è§£é‡Šæç¤ºè¯
ä¸ºä½ æ¨èå¤šå®¶é¤å…ï¼Œå¹¶è¯´æ˜æ¨èç†ç”±ã€‚ä½ å¯é€‰æ‹©â€œå–œæ¬¢/ä¸€èˆ¬/ä¸å–œæ¬¢â€ï¼Œæˆ–æŒ‡å‡ºéœ€è¦æ’é™¤/ä¼˜å…ˆçš„ç±»å‹ã€‚


````python
def get_documents(content_func=lambda row: row['name'] + '\n' + row['tag'],
                  metadata_fields=[]):
    """
    åŠ è½½å¹¶å¤„ç†é¤å…æ•°æ®ï¼Œç”Ÿæˆæ–‡æ¡£å¯¹è±¡
    
    å‚æ•°:
        content_func: å¤„ç†page_contentçš„å‡½æ•°
        metadata_fields: éœ€è¦åŒ…å«åœ¨metadataä¸­çš„å­—æ®µåˆ—è¡¨
        
    è¿”å›:
        æ–‡æ¡£å¯¹è±¡åˆ—è¡¨ï¼ŒåŒ…å«page_contentå’Œmetadata
    """
    # åŠ è½½æ•°æ®åº“
    dataset_df = pd.read_csv(DATASET_PATH)
    dataset_df.drop_duplicates(inplace=True)
    
    # å¤„ç†å†…å®¹
    dataset_df['page_content'] = dataset_df.apply(content_func, axis=1)
    
    # ç¡®ä¿page_contentåœ¨metadataä¸­
    metadata_fields = list(set(metadata_fields + ['page_content']))
    
    # ä½¿ç”¨DataFrameLoaderç”Ÿæˆæ–‡æ¡£å¯¹è±¡
    loader = DataFrameLoader(dataset_df[metadata_fields], page_content_column='page_content')
    return loader.load()

def content_func(row) -> str:
    """
    ç”Ÿæˆæ¯å®¶åº—é“ºçš„å®Œæ•´ä¿¡æ¯å­—ç¬¦ä¸²
    
    å‚æ•°:
        row: æ•°æ®è¡Œ
        
    è¿”å›:
        åŒ…å«åº—é“ºæ‰€æœ‰ä¿¡æ¯çš„å­—ç¬¦ä¸²
    """
    # åŸºç¡€ä¿¡æ¯å­—æ®µ
    content_fields = [
        "name", "address", "type", "tag", 
        "cost", "rating", "opentime_today", "opentime_week"
    ]
    
    # æ–°å¢çš„å¤šç»´è¯„ä»·å­—æ®µ
    rating_fields = [
        "dp_rating", "dp_taste_rating", "dp_env_rating",
        "dp_service_rating", "dp_comment_num"
    ]
    
    # æ¨èèœå’Œè¯„è®ºç›¸å…³å­—æ®µ
    comment_fields = [
        "dp_recommendation_dish", "dp_comment_keywords",
        "dp_top3_comments"
    ]
    
    # ç”Ÿæˆä¿¡æ¯å­—ç¬¦ä¸²
    info_parts = []
    
    # æ·»åŠ åŸºç¡€ä¿¡æ¯
    for field in content_fields:
        if pd.notna(row[field]):
            info_parts.append(f"{field}={row[field]}")
    
    # æ·»åŠ è¯„åˆ†ä¿¡æ¯
    rating_info = []
    for field in rating_fields:
        if pd.notna(row[field]):
            rating_info.append(f"{field}={row[field]}")
    if rating_info:
        info_parts.append("è¯„åˆ†ä¿¡æ¯:\n" + "\n".join(rating_info))
    
    # æ·»åŠ æ¨èèœå’Œè¯„è®º
    if pd.notna(row.get("dp_recommendation_dish")):
        info_parts.append(f"æ¨èèœ: {row['dp_recommendation_dish']}")
    
    if pd.notna(row.get("dp_comment_keywords")):
        info_parts.append(f"è¯„è®ºå…³é”®è¯: {row['dp_comment_keywords']}")
    
    if pd.notna(row.get("dp_top3_comments")):
        info_parts.append("ç²¾é€‰è¯„è®º:\n" + row['dp_top3_comments'].replace("|", "\n"))
    
    return '\n'.join(info_parts)

# è°ƒç”¨ - å¢åŠ æ–°å­—æ®µåˆ°metadataä¸­
metadata_fields = [
    "location", "opentime_week",
    "dp_rating", "dp_taste_rating", "dp_env_rating",
    "dp_service_rating", "dp_comment_num"
]

documents = get_documents(content_func, metadata_fields=metadata_fields)

# å±•ç¤ºæ›´æ–°åçš„æ–‡æ¡£ç»“æ„
print("=== æ›´æ–°åçš„æ–‡æ¡£ç¤ºä¾‹ ===")
print(documents[1].page_content)
print("\n=== å…ƒæ•°æ® ===")
print(documents[1].metadata)
````

# FAISSæ•°æ®åº“

åœ¨ä»£ç  FAISS_REVIEWS_PATH_COSINE = "faiss_index_cosine" ä¸­ï¼Œ"faiss_index_cosine" æ˜¯ FAISS å‘é‡ç´¢å¼•çš„æœ¬åœ°å­˜å‚¨è·¯å¾„ï¼Œä¿å­˜çš„æ˜¯ç»è¿‡å‘é‡åŒ–å¤„ç†åçš„æ–‡æ¡£ç´¢å¼•æ•°æ®ã€‚ä»¥ä¸‹æ˜¯è¯¦ç»†è§£é‡Šï¼š

å½“è°ƒç”¨ vector_db.save_local(FAISS_REVIEWS_PATH_COSINE) æ—¶ï¼Œä¼šåœ¨è¯¥è·¯å¾„ä¸‹ç”Ÿæˆä»¥ä¸‹æ–‡ä»¶ï¼š

äºŒè¿›åˆ¶æ–‡ä»¶ï¼Œå­˜å‚¨å‘é‡ç´¢å¼•çš„æ ¸å¿ƒæ•°æ®ï¼ˆåŒ…æ‹¬å‘é‡æ•°æ®ã€ç´¢å¼•ç»“æ„ç­‰ï¼‰ã€‚

index.pklï¼ˆå¯é€‰ï¼‰
å­˜å‚¨å…ƒæ•°æ®ï¼ˆå¦‚æ–‡æ¡£çš„åŸå§‹æ–‡æœ¬ã€IDç­‰ï¼Œéœ€é€šè¿‡ LangChain é¢å¤–é…ç½®ï¼‰ã€‚

è¿™äº›æ–‡ä»¶å…±åŒæ„æˆä¸€ä¸ªå®Œæ•´çš„å¯å¤ç”¨çš„å‘é‡æ•°æ®åº“ã€‚

````python
FAISS_REVIEWS_PATH_COSINE = "faiss_index_cosine" # å‘é‡åº“å­˜å‚¨è·¯å¾„
FAISS_INDEX_NAME = "index" # å‘é‡åº“ç´¢å¼•åç§°
FAISS_DISTANCE_STRATEGY_COSINE = "COSINE_DISTANCE" # å‘é‡åº“è·ç¦»è®¡ç®—ç­–ç•¥

# ç”¨äºæ ¹æ®csvæ•°æ®ç”Ÿæˆå‘é‡åº“çš„å‡½æ•°ã€‚documentså°±æ˜¯å‰é¢csvæ•°æ®çš„å¯¼å…¥ç»“æœã€‚embedding_modelå°±æ˜¯ä¸Šé¢å®šä¹‰çš„åµŒå…¥æ¨¡å‹ã€‚
def get_vector_database(documents, embedding_model, distance_strategy):

  vector_database = FAISS.from_documents(
      documents, embedding_model,
      distance_strategy= distance_strategy
      )
  return vector_database

# åµŒå…¥å‘é‡åº“ã€‚åˆ†æ‰¹æ¬¡å¤„ç†æ–‡æ¡£ï¼ŒåŠ å…¥äº†ç­‰å¾…æ—¶é—´ï¼Œé¿å…APIé™åˆ¶ï¼ˆç°åœ¨çš„ç‰ˆæœ¬æ˜¯æŠŠå‘é‡åº“ä¿å­˜åœ¨æœ¬åœ°ï¼Œæ²¡æœ‰å¯¹åº”é™åˆ¶ï¼‰ã€‚
import time
doclen = len(documents) # è¿™é‡Œçš„é•¿åº¦æŒ‡çš„æ˜¯å‰é¢è¯»çš„æ•°æ®çš„è¡Œæ•°ã€‚
for batch in range(doclen//100 + 1): # å°†æ¯ä¸ªåº—é“ºçš„ä¿¡æ¯ç‹¬ç«‹è½¬æ¢ä¸ºä¸€ä¸ªå‘é‡ï¼Œä¸”æ¯æ¬¡å¹¶è¡Œå¤„ç† 100 ä¸ªåº—é“ºçš„å‘é‡
    docs = documents[batch*100:(batch+1)*100]
    if batch == 0:
        vector_db = get_vector_database(docs, embedding_model, FAISS_DISTANCE_STRATEGY_COSINE)
    else:
        vector_db.merge_from(get_vector_database(docs, embedding_model, FAISS_DISTANCE_STRATEGY_COSINE))
    time.sleep(10) # æ¯æ¬¡å¤„ç†å®Œ100æ¡æ•°æ®ï¼Œä¼‘çœ 10ç§’ï¼Œé˜²æ­¢apié™åˆ¶ã€‚
    
#å‚¨å­˜å¹¶åŠ è½½å‘é‡åº“
vector_db.save_local(folder_path=FAISS_REVIEWS_PATH_COSINE, index_name=FAISS_INDEX_NAME)
vector_db = FAISS.load_local(folder_path=FAISS_REVIEWS_PATH_COSINE,
                             embeddings=embedding_model,
                             index_name=FAISS_INDEX_NAME,
                             allow_dangerous_deserialization=True) # å…è®¸ååºåˆ—åŒ–
````

````python
def create_user_preference(
    environment=3, 
    taste=4,
    service=3,
    cost_performance=4,
    hygiene=4,
    health=3,
    waiting_time=2,
    distance=3,
    preferred_cuisines=["å·èœ", "ç«é”…"],
    disliked_cuisines=["æ—¥æ–™"],
    budget_range=[50, 150],
    special_requirements="æ— "
):
    """
    åˆ›å»ºç”¨æˆ·åå¥½å­—å…¸
    """
    return {
        "ç¯å¢ƒ": environment,
        "å£å‘³": taste,
        "æœåŠ¡": service,
        "æ€§ä»·æ¯”": cost_performance,
        "å«ç”Ÿ": hygiene,
        "è¥å…»å¥åº·": health,
        "æ’é˜Ÿæ—¶é—´": waiting_time,
        "è·ç¦»": distance,
        "åå¥½èœç³»": preferred_cuisines,
        "ä¸å–œæ¬¢çš„èœç³»": disliked_cuisines,
        "é¢„ç®—èŒƒå›´": budget_range,
        "ç‰¹æ®Šè¦æ±‚": special_requirements
    }

system_prompt_template = """
# ä½ çš„è§’è‰²

ä½ æ˜¯â€œèœæ ¹æ¢â€â€”â€”ä¸€åæ™ºèƒ½ç¾é£Ÿæ¨èåŠ©æ‰‹ã€‚ä½ çš„å·¥ä½œæ˜¯æ ¹æ®ç”¨æˆ·çš„ä¸ªæ€§åŒ–åå¥½å’Œå®æ—¶éœ€æ±‚ï¼Œä»æ•°æ®åº“ä¸­æ¨èæœ€ä¼˜é¤å…ã€‚

# 1. ç”¨æˆ·ä¸ªäººåå¥½

ç”¨æˆ·çš„ä¸ªäººåå¥½å¦‚ä¸‹ï¼ˆå·²ç”±å‰ç«¯é¡µé¢é‡‡é›†ï¼‰ï¼š
'''
{user_preference}
'''
- åå¥½æŒ‡æ ‡åŒ…æ‹¬ï¼šç¯å¢ƒã€å£å‘³ã€æœåŠ¡ã€æ€§ä»·æ¯”ã€å«ç”Ÿã€è¥å…»å¥åº·ã€æ’é˜Ÿæ—¶é—´ã€è·ç¦»ç­‰ã€‚
- æ¯é¡¹å¾—åˆ†ä¸º0-5åˆ†ï¼Œæ•°å€¼è¶Šé«˜ä»£è¡¨ç”¨æˆ·åœ¨æŒ‘é€‰é¤å…æ—¶å¯¹è¯¥é¡¹è¦æ±‚è¶Šé«˜ï¼ŒæŒ‡æ ‡è¶Šé å‰è¯´æ˜è¶Šé‡è¦ã€‚
- è¯·ç»“åˆè¿™äº›åˆ†å€¼ï¼Œä¸ºåç»­é¤å…ç­›é€‰ä¸åŠ æƒæ‰“åˆ†åˆ†é…ä¸åŒæƒé‡ï¼Œä¼˜å…ˆæ»¡è¶³å¾—åˆ†é«˜å’Œæ’åé å‰çš„æŒ‡æ ‡é¡¹ã€‚
- åå¥½å«ä¹‰å‚è€ƒï¼š  
    - ç¯å¢ƒï¼šé‡è§†é¤å…ç¯å¢ƒã€æ°›å›´ã€å®‰é™åº¦ã€ç©ºé—´èˆ’é€‚åº¦  
    - å£å‘³ï¼šé‡è§†èœå“é£å‘³ã€è¾£åº¦ã€ç”œåº¦ã€é£Ÿææ­£å®—æ€§  
    - æœåŠ¡ï¼šé‡è§†æœåŠ¡æ€åº¦ã€æ•ˆç‡ã€äººå‘˜ç´ è´¨  
    - æ€§ä»·æ¯”ï¼šè¿½æ±‚èŠ±æ›´å°‘çš„é’±äº«å—æ›´å¥½æœåŠ¡å’Œèœå“  
    - å«ç”Ÿï¼šå¯¹ç”¨é¤å«ç”Ÿå’Œé£Ÿæå®‰å…¨çš„é‡è§†  
    - è¥å…»å¥åº·ï¼šè¿½æ±‚å¥åº·ã€ä½æ²¹ä½ç›ã€è¥å…»æ­é…  
    - æ’é˜Ÿæ—¶é—´ï¼šå…³æ³¨æ˜¯å¦éœ€ç­‰ä½ã€å‡ºé¤é€Ÿåº¦  
    - è·ç¦»ï¼šä¼˜å…ˆè€ƒè™‘è·ç¦»è¿‘ã€äº¤é€šä¾¿åˆ©çš„é¤å…
    è¯·ç‰¹åˆ«æ³¨æ„ä»¥ä¸‹å…³é”®ä¿¡æ¯ï¼š
    - å„ç»´åº¦è¯„åˆ†(0-5åˆ†): {preference_scores}
    - åå¥½èœç³»: {preferred_cuisines}
    - ä¸å–œæ¬¢çš„èœç³»: {disliked_cuisines}
    - é¢„ç®—èŒƒå›´: {budget_range}å…ƒ
    - ç‰¹æ®Šè¦æ±‚: {special_requirements}

    # 2. æ¨èç­–ç•¥
    - ä¼˜å…ˆåŒ¹é…ç”¨æˆ·è¯„åˆ†é«˜çš„ç»´åº¦(4-5åˆ†)
    - ç¡®ä¿æ¨èé¤å…åœ¨ç”¨æˆ·é¢„ç®—èŒƒå›´å†…
    - é¿å…æ¨èç”¨æˆ·ä¸å–œæ¬¢çš„èœç³»
    - è€ƒè™‘ç”¨æˆ·ç‰¹æ®Šè¦æ±‚
# 2. ç”¨æˆ·å®æ—¶éœ€æ±‚å’Œåœºæ™¯é‡‡é›†

-ç”¨æˆ·ä¼šç”¨è‡ªç„¶è¯­è¨€è¡¨è¾¾æœ¬æ¬¡ç”¨é¤éœ€æ±‚ï¼ˆå¦‚é¢„ç®—ã€äººæ•°ã€è·ç¦»ã€æœŸæœ›åœºæ™¯/èœç³»/å£å‘³ç­‰ï¼‰ï¼Œä½ éœ€è¦æ™ºèƒ½è§£æè¿™äº›å†…å®¹ï¼Œå°†å…¶ä¸ä¸ªäººåå¥½ç»“åˆèµ·æ¥ï¼Œä¸ºç”¨æˆ·è¿›è¡ŒåŒ¹é…æ¨èã€‚
-é¼“åŠ±ç”¨æˆ·ç”¨è‡ªç„¶è¯­è¨€è¡¨è¾¾å½“å‰ç”¨é¤éœ€æ±‚ï¼ˆå¦‚é¢„ç®—ã€äººæ•°ã€è·ç¦»ã€ç±»å‹ã€èœç³»ã€æœŸæœ›åœºæ™¯/æ°›å›´ç­‰ï¼‰ï¼Œæ”¯æŒç”¨æˆ·è‡ªç”±æè¿°ï¼Œä¹Ÿå¯è¿½é—®è¡¥å…¨å…³é”®ä¿¡æ¯ã€‚
-ä½ éœ€è§£æè¿™äº›éœ€æ±‚ï¼Œæ˜ç¡®ç¡¬æ€§ï¼ˆé¢„ç®—/è¥ä¸š/è·ç¦»/äººæ•°/æ—¶æ®µï¼‰ä¸è½¯æ€§ï¼ˆå£å‘³/æ°›å›´/å¥åº·/ç¯å¢ƒï¼‰çº¦æŸã€‚

# 3. é¤å…æ•°æ®åº“å­—æ®µè¯´æ˜

ä½ æ‰€å‚è€ƒçš„æ•°æ®åº“ä¸­ï¼Œæ¯å®¶é¤å…åŒ…å«ä»¥ä¸‹å­—æ®µï¼š
- nameï¼ˆé¤å…åç§°ï¼‰
- dp_costï¼ˆå¤§ä¼—ç‚¹è¯„äººå‡æ¶ˆè´¹ï¼‰
- dp_ratingï¼ˆå¤§ä¼—ç‚¹è¯„ç»¼åˆè¯„åˆ†ï¼‰
- dp_taste_ratingï¼ˆå¤§ä¼—ç‚¹è¯„å£å‘³è¯„åˆ†ï¼‰
- dp_env_ratingï¼ˆå¤§ä¼—ç‚¹è¯„ç¯å¢ƒè¯„åˆ†ï¼‰
- dp_service_ratingï¼ˆå¤§ä¼—ç‚¹è¯„æœåŠ¡è¯„åˆ†ï¼‰
- dp_comment_numï¼ˆå¤§ä¼—ç‚¹è¯„è¯„è®ºæ•°ï¼‰
- dp_recommendation_dishï¼ˆæ¨èèœå“ï¼‰
- dp_comment_keywordsï¼ˆé«˜é¢‘è¯„è®ºå…³é”®è¯ï¼‰
- dp_top3_commentsï¼ˆç²¾é€‰è¯„è®ºï¼‰
- addressï¼ˆé—¨åº—åœ°å€ï¼‰
- locationï¼ˆåœ°ç†åæ ‡æˆ–ä½ç½®æè¿°ï¼‰
- typeï¼ˆé¤å…ç±»å‹/èœç³»ï¼‰
- telï¼ˆè”ç³»ç”µè¯ï¼‰
- costï¼ˆå…¶ä»–å¹³å°äººå‡æ¶ˆè´¹ï¼‰
- ratingï¼ˆå¹³å°ç»¼åˆè¯„åˆ†ï¼‰
- opentime_todayï¼ˆä»Šæ—¥è¥ä¸šæ—¶é—´ï¼‰
- opentime_weekï¼ˆæ¯å‘¨è¥ä¸šæ—¶é—´ï¼‰
- tagï¼ˆæ ‡ç­¾/ç‰¹è‰²ï¼‰
è¯·æ³¨æ„ä»¥ä¸Šä¿¡æ¯å¯èƒ½ä¼šæœ‰ç¡®å®æˆ–ä¸å®Œæ•´çš„æƒ…å†µï¼Œéœ€è¦ä½ å†ç»“åˆç½‘ç»œæœç´¢ç»™å‡ºæ›´åŠ å…¨é¢å®¢è§‚çš„ç»“æœï¼Œå¦‚æœæ²¡æœ‰è·å–åˆ°æ¸…æ™°æ˜ç¡®çš„ä¿¡æ¯å¯å¿½ç•¥è¯¥ç»´åº¦ï¼Œä¸è¦è¾“å‡ºè™šæ„å†…å®¹ã€‚

# 4. æ¨èè¾“å‡ºè¦æ±‚
æ¯è½®æ¨èè¯·ä¸¥æ ¼è¾“å‡ºï¼š
- æŒ‰â€œä¸ªäººåå¥½+å®æ—¶éœ€æ±‚â€ç»¼åˆå¾—åˆ†æ’åºï¼Œæ¨è3~5å®¶æœ€ä¼˜é¤å…ã€‚
- æ¯å®¶é¤å…è¯¦ç»†è¾“å‡ºï¼š
    1. åŸºç¡€ä¿¡æ¯ï¼šåç§°ã€åœ°å€ã€ç±»å‹/èœç³»ã€æ ‡ç­¾/ç‰¹è‰²ã€è”ç³»ç”µè¯ã€äººå‡æ¶ˆè´¹ã€è¥ä¸šæ—¶é—´ã€è·ç¦»ç­‰
    2. å¤šç»´åŒ¹é…æ‰“åˆ†ï¼šç»¼åˆå¾—åˆ†ï¼ˆ0-100åˆ†/5åˆ†åˆ¶ï¼‰ï¼Œå¹¶å¯¹å„ä¸»ç»´åº¦ï¼ˆç¯å¢ƒã€å£å‘³ã€æœåŠ¡ã€æ€§ä»·æ¯”ç­‰ï¼‰å•ç‹¬æ‰“åˆ†ï¼Œæ¸…æ™°å±•ç¤ºä¸ç”¨æˆ·åå¥½å¯¹åº”åˆ†å€¼ï¼Œå¹¶è¯´æ˜åŸå› 
    3. AIç”Ÿæˆä¸ªæ€§åŒ–æ¨èç†ç”±ï¼š**ç»“åˆç”¨æˆ·é«˜æƒé‡æŒ‡æ ‡å…·ä½“å±•å¼€**ï¼Œå¦‚â€œè¯¥åº—ç¯å¢ƒè¯„åˆ†4.8ï¼Œå®‰é™ä¼˜é›…ï¼Œæåº¦é€‚åˆä½ å¯¹ç¯å¢ƒçš„é«˜è¦æ±‚â€ï¼Œå¹¶é€‚å½“å¼•ç”¨ç²¾é€‰è¯„è®ºã€æ¨èèœã€å…³é”®è¯
    4. æœ€ä½³äº¤é€šæ–¹å¼ä¸æ‰€éœ€æ—¶é—´ï¼šè¯·æ ¹æ®ç”¨æˆ·å‡ºå‘åœ°ï¼ˆå¦‚æœªè¾“å…¥ï¼Œè¯·ä¸»åŠ¨å¼•å¯¼ç”¨æˆ·è¡¥å……è‡ªå·±çš„å…·ä½“ä½ç½®ï¼‰ï¼Œä¸ºæ¯å®¶é¤å…ç»™å‡ºæ¨èçš„äº¤é€šæ–¹å¼ï¼ˆæ­¥è¡Œã€åœ°é“ã€æ‰“è½¦ã€å…¬äº¤ç­‰ï¼‰ä¸å¤§è‡´æ‰€éœ€æ—¶é—´ï¼ˆå¦‚æ­¥è¡Œ10åˆ†é’Ÿã€åœ°é“2ç«™å…±20åˆ†é’Ÿç­‰ï¼‰ï¼Œæå‡å®é™…å¯è¾¾æ€§ä¸ç”¨æˆ·ä½“éªŒã€‚
    5. ç”¨æˆ·è¡¥å……å¼•å¯¼ï¼šä¸»åŠ¨å»ºè®®ç”¨æˆ·è¿›ä¸€æ­¥ç»†åŒ–éœ€æ±‚ï¼ˆå¦‚é¢„ç®—ã€å£å‘³ã€åœºæ™¯ã€ç‰¹æ®Šè¦æ±‚ã€å‡ºå‘åœ°ç­‰ï¼‰ï¼Œå¦‚æœæ¨èæœªå®Œå…¨åŒ¹é…é«˜åˆ†é¡¹ï¼Œè¯·æç¤ºå¯è¡¥å……æ›´å¤šé™åˆ¶æ¡ä»¶ä»¥ä¼˜åŒ–ç»“æœã€‚ç‰¹åˆ«æ˜¯åœ¨æœªè·å¾—ç”¨æˆ·å…·ä½“ä½ç½®æ—¶ï¼Œæé†’å…¶è¡¥å……å‡ºå‘åœ°ä»¥ä¼˜åŒ–è·¯çº¿æ¨èã€‚
# è¡¥å……è¯´æ˜
- å¦‚æœç”¨æˆ·æœªæä¾›å‡ºå‘åœ°ï¼Œè¯·åœ¨å›å¤æœ«å°¾ç¤¼è²Œå»ºè®®ï¼šâ€œä¸ºè·å¾—æ›´å‡†ç¡®çš„äº¤é€šæ–¹æ¡ˆï¼Œè¯·å‘ŠçŸ¥ä½ å½“å‰æˆ–å¸Œæœ›å‡ºå‘çš„å…·ä½“ä½ç½®â€ã€‚
- äº¤é€šæ–¹æ¡ˆå¯æ ¹æ®é¤å…åœ°å€å’Œç”¨æˆ·ä½ç½®ç”Ÿæˆï¼ˆå¦‚æ— æ³•è·å–å®é™…è·¯çº¿ï¼Œå¯ç”¨â€œé¢„è®¡æ­¥è¡Œçº¦Xåˆ†é’Ÿâ€æ ¼å¼ä¼°ç®—ï¼‰ã€‚


# è§„åˆ™
- ä½ åªèƒ½åŸºäºæ•°æ®åº“ç°æœ‰ä¿¡æ¯æ¨èï¼Œä¸å…è®¸è™šæ„ä¸å­˜åœ¨çš„é¤å…ã€‚
- å¦‚æŸé¡¹ä¿¡æ¯ç¼ºå¤±ï¼Œè¯·å¦‚å®è¯´æ˜â€œè¯¥é¡¹æš‚æ— æ•°æ®â€ã€‚
- éé¤é¥®ã€æ— å…³é—®é¢˜è¯·å§”å©‰å›å¤â€œä»…èƒ½ä¸ºæ‚¨æä¾›ç¾é£Ÿ/é¤å…æ¨èæœåŠ¡â€ã€‚
- æ¨èå†…å®¹éœ€ç»“æ„åŒ–ã€æ¡ç†æ¸…æ™°ã€æ˜“äºç”¨æˆ·ç†è§£å’Œå†³ç­–ã€‚


# å½“å‰å¯¹è¯å†å²
{history}

# æ•°æ®åº“å†…å®¹
{context}
"""

def format_user_preference(user_pref):
    """æ ¼å¼åŒ–ç”¨æˆ·åå¥½ä¸ºå­—ç¬¦ä¸²"""
    scores = "\n".join([f"{k}: {v}åˆ†" for k,v in user_pref.items() if isinstance(v, int)])
    # å¥å£®æ€§å¤„ç†é¢„ç®—èŒƒå›´
    budget_range = user_pref.get('é¢„ç®—èŒƒå›´', [])
    if isinstance(budget_range, list) and len(budget_range) >= 2:
        budget_str = f"{budget_range[0]}-{budget_range[1]}å…ƒ"
    elif isinstance(budget_range, list) and len(budget_range) == 1:
        budget_str = f"{budget_range[0]}å…ƒ"
    else:
        budget_str = "æœªå¡«å†™"
    return (
        f"ç”¨æˆ·åå¥½è¯„åˆ†:\n{scores}\n\n"
        f"åå¥½èœç³»: {', '.join(user_pref.get('åå¥½èœç³»', []))}\n"
        f"ä¸å–œæ¬¢çš„èœç³»: {', '.join(user_pref.get('ä¸å–œæ¬¢çš„èœç³»', []))}\n"
        f"é¢„ç®—èŒƒå›´: {budget_str}\n"
        f"ç‰¹æ®Šè¦æ±‚: {user_pref.get('ç‰¹æ®Šè¦æ±‚', 'æ— ')}"
    )

def extract_preference_vars(user_pref):
    # preference_scores: ç¯å¢ƒã€å£å‘³ã€æœåŠ¡ã€æ€§ä»·æ¯”ã€å«ç”Ÿã€è¥å…»å¥åº·ã€æ’é˜Ÿæ—¶é—´ã€è·ç¦»
    scores = []
    for k in ["ç¯å¢ƒ", "å£å‘³", "æœåŠ¡", "æ€§ä»·æ¯”", "å«ç”Ÿ", "è¥å…»å¥åº·", "æ’é˜Ÿæ—¶é—´", "è·ç¦»"]:
        if k in user_pref:
            scores.append(f"{k}: {user_pref[k]}åˆ†")
    return {
        "preference_scores": "\n".join(scores),
        "preferred_cuisines": ", ".join(user_pref.get("åå¥½èœç³»", [])),
        "disliked_cuisines": ", ".join(user_pref.get("ä¸å–œæ¬¢çš„èœç³»", [])),
        "budget_range": f"{user_pref.get('é¢„ç®—èŒƒå›´', [0,0])[0]}-{user_pref.get('é¢„ç®—èŒƒå›´', [0,0])[1]}å…ƒ",
        "special_requirements": user_pref.get("ç‰¹æ®Šè¦æ±‚", "æ— "),
    }
````

````python
from langchain.prompts import (
    PromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
    ChatPromptTemplate,
)
from langchain_core.output_parsers import StrOutputParser
from langchain_community.vectorstores import FAISS
from langchain.schema.runnable import RunnablePassthrough, RunnableMap, RunnableLambda
from langchain.memory import ConversationBufferMemory

# å‡è®¾ vector_db å’Œ llm å·²æå‰å®šä¹‰
# vector_db = FAISS.load_local(...)
# llm = ...

# åˆå§‹åŒ–å¯¹è¯å†å²è®°å¿†æ¨¡å—
memory = ConversationBufferMemory(memory_key="history", return_messages=True)

human_prompt_template = """
{question}
user_preference={user_preference}
"""

# ä½¿ç”¨langchainè‡ªå¸¦çš„promptsæ¨¡æ¿ç¼–è¾‘å™¨æ¥ç¼–è¾‘promptæ ¼å¼
system_prompt = SystemMessagePromptTemplate(
    prompt=PromptTemplate(
        input_variables=["history", "context"], template=system_prompt_template
    )
)
human_prompt = HumanMessagePromptTemplate(
    prompt=PromptTemplate(input_variables=["question"], template=human_prompt_template)
)
messages = [system_prompt, human_prompt]
total_prompt_template = ChatPromptTemplate(
    input_variables=[
        "history", "context", "question", "user_preference",
        "preference_scores", "preferred_cuisines", "disliked_cuisines", "budget_range", "special_requirements"
    ],
    messages=messages,
)

# é…ç½®RAGçš„åˆæ­¥æ£€ç´¢å™¨
# ä¿®æ­£ï¼šç¡®ä¿åªå°† question å­—æ®µä¼ é€’ç»™ retriever
reviews_retriever = RunnableLambda(lambda x: x["question"]) | vector_db.as_retriever(search_kwargs={'k': 20,})

# é…ç½®å’Œå¤§æ¨¡å‹äº¤äº’çš„å®Œæ•´chatbotï¼ŒåŠ å…¥å†å²è®°å¿†
# ç¡®ä¿ user_preference å·²å®šä¹‰ï¼ˆå¦‚æœªå®šä¹‰åˆ™åˆå§‹åŒ–ä¸ºç©ºå­—å…¸æˆ–é‡‡é›†ç”¨æˆ·ç”»åƒï¼‰
try:
    user_preference
except NameError:
    user_preference = {}

review_chain = (
    RunnableMap({
        "history": RunnablePassthrough(lambda _: memory.load_memory_variables({})),
        "context": reviews_retriever,  # ä¿®æ­£å context åªä¼ é€’ question å­—æ®µ
        "question": RunnablePassthrough(),
        "user_preference": lambda x: format_user_preference(user_preference),
        "preference_scores": lambda x: extract_preference_vars(user_preference)["preference_scores"],
        "preferred_cuisines": lambda x: extract_preference_vars(user_preference)["preferred_cuisines"],
        "disliked_cuisines": lambda x: extract_preference_vars(user_preference)["disliked_cuisines"],
        "budget_range": lambda x: extract_preference_vars(user_preference)["budget_range"],
        "special_requirements": lambda x: extract_preference_vars(user_preference)["special_requirements"],
    })
    | total_prompt_template
    | llm
    | StrOutputParser()
)

# === ç”¨æˆ·ç”»åƒé‡‡é›†ä¸è½¬æ¢ ===
def collect_user_profile():
    """
    äº¤äº’å¼é‡‡é›†ç”¨æˆ·ç”»åƒä¿¡æ¯ï¼Œå¹¶è¿”å›ç»Ÿä¸€æ ¼å¼çš„ç”¨æˆ·åå¥½å­—å…¸
    """
    print("è¯·æ ¹æ®æç¤ºè¾“å…¥ä½ çš„ç”¨é¤åå¥½ï¼ˆ1-5åˆ†ï¼Œ5åˆ†æœ€é‡è§†ï¼‰")
    def ask_score(q):
        while True:
            try:
                s = int(input(q))
                if 1 <= s <= 5:
                    return s
                else:
                    print("è¯·è¾“å…¥1-5ä¹‹é—´çš„æ•´æ•°")
            except:
                print("è¯·è¾“å…¥æ•°å­—")
    cost_performance = ask_score("è¯·ç”¨1-5åˆ†è¯„ä»·ä½ å¯¹â€˜æ€§ä»·æ¯”â€™çš„é‡è§†ç¨‹åº¦ï¼ˆ1-5åˆ†ï¼‰ï¼š")
    hygiene = ask_score("è¯·ç”¨1-5åˆ†è¯„ä»·ä½ å¯¹â€˜å«ç”Ÿâ€™çš„é‡è§†ç¨‹åº¦ï¼ˆ1-5åˆ†ï¼‰ï¼š")
    taste = ask_score("è¯·ç”¨1-5åˆ†è¯„ä»·ä½ å¯¹â€˜å£å‘³â€™çš„é‡è§†ç¨‹åº¦ï¼ˆ1-5åˆ†ï¼‰ï¼š")
    service = ask_score("è¯·ç”¨1-5åˆ†è¯„ä»·ä½ å¯¹â€˜æœåŠ¡â€™çš„é‡è§†ç¨‹åº¦ï¼ˆ1-5åˆ†ï¼‰ï¼š")
    waiting_time = ask_score("è¯·ç”¨1-5åˆ†è¯„ä»·ä½ å¯¹â€˜æ’é˜Ÿæ—¶é—´â€™çš„é‡è§†ç¨‹åº¦ï¼ˆ1-5åˆ†ï¼‰ï¼š")
    health = ask_score("è¯·ç”¨1-5åˆ†è¯„ä»·ä½ å¯¹â€˜è¥å…»å¥åº·â€™çš„é‡è§†ç¨‹åº¦ï¼ˆ1-5åˆ†ï¼‰ï¼š")
    environment = ask_score("è¯·ç”¨1-5åˆ†è¯„ä»·ä½ å¯¹â€˜ç¯å¢ƒæ°›å›´â€™çš„é‡è§†ç¨‹åº¦ï¼ˆ1-5åˆ†ï¼‰ï¼š")
    distance = ask_score("è¯·ç”¨1-5åˆ†è¯„ä»·ä½ å¯¹â€˜è·ç¦»è¿œè¿‘â€™çš„é‡è§†ç¨‹åº¦ï¼ˆ1-5åˆ†ï¼‰ï¼š")
    preferred_cuisines = input("ä½ åå¥½çš„èœç³»æœ‰å“ªäº›ï¼Ÿï¼ˆç”¨é€—å·åˆ†éš”ï¼Œå¦‚å·èœ,ç«é”…ï¼Œå¯è·³è¿‡ï¼‰ï¼š").split(",") if input("æ˜¯å¦æœ‰åå¥½èœç³»ï¼Ÿ(y/n):").lower() == "y" else []
    disliked_cuisines = input("ä½ ä¸å–œæ¬¢çš„èœç³»æœ‰å“ªäº›ï¼Ÿï¼ˆç”¨é€—å·åˆ†éš”ï¼Œå¦‚æ—¥æ–™ï¼Œå¯è·³è¿‡ï¼‰ï¼š").split(",") if input("æ˜¯å¦æœ‰ä¸å–œæ¬¢çš„èœç³»ï¼Ÿ(y/n):").lower() == "y" else []
    budget_str = input("ä½ çš„é¢„ç®—èŒƒå›´æ˜¯å¤šå°‘ï¼Ÿï¼ˆå¦‚30-80ï¼Œå•ä½å…ƒï¼‰ï¼š")
    try:
        budget_range = [int(x) for x in budget_str.split("-")]
    except:
        budget_range = [0, 999]
    special_requirements = input("ä½ è¿˜æœ‰ä»€ä¹ˆç‰¹åˆ«çš„é¥®é£Ÿåå¥½ã€å¿Œå£ã€ç¦å¿Œæˆ–ç”¨é¤ä¹ æƒ¯éœ€è¦è¡¥å……å—ï¼Ÿï¼ˆå¦‚æ— å¯è·³è¿‡ï¼‰ï¼š") or "æ— "
    # ç»Ÿä¸€æ ¼å¼
    user_pref = {
        "ç¯å¢ƒ": environment,
        "å£å‘³": taste,
        "æœåŠ¡": service,
        "æ€§ä»·æ¯”": cost_performance,
        "å«ç”Ÿ": hygiene,
        "è¥å…»å¥åº·": health,
        "æ’é˜Ÿæ—¶é—´": waiting_time,
        "è·ç¦»": distance,
        "åå¥½èœç³»": [c.strip() for c in preferred_cuisines if c.strip()],
        "ä¸å–œæ¬¢çš„èœç³»": [c.strip() for c in disliked_cuisines if c.strip()],
        "é¢„ç®—èŒƒå›´": budget_range,
        "ç‰¹æ®Šè¦æ±‚": special_requirements
    }
    print("ã€ä½ çš„ç”¨æˆ·ç”»åƒå·²ç”Ÿæˆã€‘\n", user_pref)
    return user_pref

# === ä¸»æµç¨‹ç¤ºä¾‹ ===
def main():
    print("æ¬¢è¿ä½¿ç”¨æ™ºèƒ½ç¾é£Ÿæ¨èç³»ç»Ÿï¼")
    # 1. é‡‡é›†ç”¨æˆ·ç”»åƒ
    user_profile = collect_user_profile()
    # 2. è¿›å…¥æ¨èé—®ç­”å¾ªç¯
    from langchain.memory import ConversationBufferMemory
    memory = ConversationBufferMemory(memory_key="history", return_messages=True)
    print("\nè¯·è¾“å…¥ä½ çš„æœ¬æ¬¡ç”¨é¤éœ€æ±‚ï¼ˆå¦‚â€˜é¢„ç®—50å…ƒå†…ï¼Œæƒ³åƒè¾£çš„ï¼Œé€‚åˆ2äººï¼Œè·ç¦»ä¸è¿œâ€™ï¼‰ï¼š")
    while True:
        question = input("ä½ çš„ç”¨é¤éœ€æ±‚/é—®é¢˜ï¼š")
        if not question.strip():
            print("è¾“å…¥ä¸ºç©ºï¼Œå·²é€€å‡ºã€‚")
            break
        # æ„é€ è¾“å…¥
        input_data = {
            "question": question,
            "user_preference": format_user_preference(user_profile)
        }
        # æ¨è
        response = review_chain.invoke(input_data)
        # ä¿å­˜ä¸Šä¸‹æ–‡
        memory.save_context({"input": question}, {"output": response})
        print("\n=== æ¨èç»“æœ ===")
        print(response)
        again = input("\næ˜¯å¦ç»§ç»­æé—®ï¼Ÿ(yç»§ç»­ï¼Œå…¶ä»–é€€å‡º)ï¼š")
        if again.lower() != "y":
            break

# === å®Œæ•´æµç¨‹ç¤ºä¾‹ ===
if __name__ == "__main__":
    main()
````
